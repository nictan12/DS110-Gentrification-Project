```python
'''We decided that the dataset that we had didn't give us all the information that we needed, so we decided to create one excel sheet with
all the information that we've attained while also adding metro average values which means that we're also taking into account the surrounding areas 
around NYC'''
import pandas as pd
from google.colab import files

uploaded = files.upload()

TestData = pd.DataFrame(pd.read_csv('TestData6.csv'))
#When doing machine learning, I wasn't able to run it correctly because there were NaN's, but was able to change by filling NaN's with 0
TestData.fillna(0, inplace = True)
```

```python
#These three evaluate methods gives me three new columns that calculate for rate of change and difference within the data
TestData_merge = TestData.eval('Change_NMHV = (mhv2020 - mhv2000)/mhv2000')
Next_Data = TestData_merge.eval('Difference_NMHI = mhhinc2020 - mhhinc2000')
Super_Data = Next_Data.eval('Change_MMHV = (MMHV2020 - MMHV2000)/MMHV2000')

'''The model works on using metro values as the benchmark and comparing those to the respective neighborhood values.
N(MHV) stands for Neighborhood (Median Housing Value);
NMHI stands for Neighborhood Median Household Income;
MHHINC stands for Median Household Income (it's the acronym used in the models);
MMHV stands for Metro Median Household Value'''

Super_Data
```

```python
#All data types have to be the same if I want to use Boolean expressions to check for gentrification
Mega_Data['Neighborhood Median Household Income 2020'] = Mega_Data['Neighborhood Median Household Income 2020'].astype(float)
Mega_Data['Neighborhood Median Home Value 2020'] = Mega_Data['Neighborhood Median Home Value 2020'].astype(float)
Mega_Data['Neighborhood Median Gross Rent 2020'] = Mega_Data['Neighborhood Median Gross Rent 2020'].astype(float)
Mega_Data['Neighborhood Median Year Built 2020'] = Mega_Data['Neighborhood Median Year Built 2020'].astype(float)
Mega_Data['Neighborhood Percent of Units Built in 1939 or Before 2020'] = Mega_Data['Neighborhood Percent of Units Built in 1939 or Before 2020'].astype(float)
Mega_Data['Neighborhood People with at Least a Bachelors Degree 2020'] = Mega_Data['Neighborhood People with at Least a Bachelors Degree 2020'].astype(float)
Mega_Data['Neighborhood Percent Black Population 2020'] = Mega_Data['Neighborhood Percent Black Population 2020'].astype(float)
Mega_Data['Neighborhood Median Household Income 2000'] = Mega_Data['Neighborhood Median Household Income 2000'].astype(float)
Mega_Data['Neighborhood Median Home Value 2000'] = Mega_Data['Neighborhood Median Home Value 2000'].astype(float)
Mega_Data['Neighborhood Median Gross Rent 2010'] = Mega_Data['Neighborhood Median Gross Rent 2010'].astype(float)
Mega_Data['Neighborhood Median Year Built 2000'] = Mega_Data['Neighborhood Median Year Built 2000'].astype(float)
Mega_Data['Neighborhood Percent of Units Built in 1939 or Before 2000'] = Mega_Data['Neighborhood Percent of Units Built in 1939 or Before 2000'].astype(float)
Mega_Data['Neighborhood People with at Least a Bachelors Degree 2000'] = Mega_Data['Neighborhood People with at Least a Bachelors Degree 2000'].astype(float)
Mega_Data['Neighborhood Percent Black Population 2000'] = Mega_Data['Neighborhood Percent Black Population 2000'].astype(float)
Mega_Data['Metro Median Bachelors Degree 2020'] = Mega_Data['Metro Median Bachelors Degree 2020'].astype(float)
Mega_Data['Metro Median Home Value 2000'] = Mega_Data['Metro Median Home Value 2000'].astype(float)
Mega_Data['Metro Median Home Value 2020'] = Mega_Data['Metro Median Home Value 2020'].astype(float)
Mega_Data['Metro Median Household Income 2000'] = Mega_Data['Metro Median Household Income 2000'].astype(float)
Mega_Data['Metro Median Household Income 2020'] = Mega_Data['Metro Median Household Income 2020'].astype(float)
Mega_Data['Metro Median Year Built 2000'] = Mega_Data['Metro Median Year Built 2000'].astype(float)
Mega_Data['Metro Median Year Built 2020'] = Mega_Data['Metro Median Year Built 2020'].astype(float)
Mega_Data['Metro Median Household Income 2000'] = Mega_Data['Metro Median Household Income 2000'].astype(float)
Mega_Data['Metro Median Household Income 2020'] = Mega_Data['Metro Median Household Income 2020'].astype(float)
Mega_Data['Metro Median Year Built 2000'] = Mega_Data['Metro Median Year Built 2000'].astype(float)
Mega_Data['Metro Median Year Built 2020'] = Mega_Data['Metro Median Year Built 2020'].astype(float)
#Re-checking if all data types are the same
Mega_Data.dtypes
#This qualified column determines if the NYC neighborhood qualifies as gentrified using the Freeman model
```

```python
#This qualification variable determines if the NYC neighborhood qualifies in determining gentrification using the various gentrification models
model_A_qualification = (Mega_Data['Neighborhood Median Household Income 2000'] < Mega_Data['Metro Median Household Income 2000']) & (Mega_Data['Neighborhood Median Household Income 2020'] < Mega_Data['Metro Median Household Income 2020']) & (Mega_Data['Neighborhood Median Year Built 2000'] < Mega_Data['Metro Median Year Built 2000']) & (Mega_Data['Neighborhood Median Year Built 2020'] < Mega_Data['Metro Median Year Built 2020'])
#This function figures out if the NYC neighborhood is gentrified or not by going through the three models to check if it passes the condition for gentrification
def gentrified(Mega_Data):
  #Boolean expression for Freeman model conditions
  model_A = model_A_qualification & (Mega_Data['Neighborhood People with at Least a Bachelors Degree 2020'] > Mega_Data['Metro Median Bachelors Degree 2020']) & (Mega_Data['Change_NMHV'] > Mega_Data['Change_MMHV'])
  #Boolean expression for McKinnish model conditions
  model_B = (Mega_Data['Neighborhood Median Household Income 2020'] <= Mega_Data['Metro Median Household Income 2020'].quantile(0.2)) & (Mega_Data['Difference_NMHI'] >= 20000)
  #Boolean expression for Ellen & O'Regan model conditions
  model_C = ((Mega_Data['Neighborhood Median Household Income 2000']/Mega_Data['Metro Median Household Income 2000'] < 0.7) & ((((Mega_Data['Neighborhood Median Household Income 2020']/Mega_Data['Metro Median Household Income 2020']) - (Mega_Data['Neighborhood Median Household Income 2000']/Mega_Data['Metro Median Household Income 2000'])) / (Mega_Data['Neighborhood Median Household Income 2000'] / Mega_Data['Metro Median Household Income 2000'])) >= 0.1))
  #Counts how many models each row has passed
  count = model_A.astype(int) + model_B.astype(int) + model_C.astype(int)
  Mega_Data['Gentrified'] = count 
  #For a NYC neighborhood to pass as gentrified, it has to pass at least two of the models that we have provided
  gentrified = Mega_Data[Mega_Data['Gentrified'] >= 0]
  return gentrified
#Outputs each row with gentrification labels from the amount of models each row passed
gentrified(Mega_Data)
```
